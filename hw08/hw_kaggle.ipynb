Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

```python
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self, X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        X_copy = X.copy()
        if self.columns is None:
            for colname, col in X_copy.iteritems():
                X_copy[colname] = LabelEncoder().fit_transform(col)
        else:
            for col in self.columns:
                X_copy[col] = LabelEncoder().fit_transform(X_copy[col])
        return X_copy

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

```python
# YOUR CODE
df = data.copy()

# scale numerical
scaler = StandardScaler()
scaler.fit(df[num_cols])
df[num_cols] = scaler.transform(df[num_cols])

# one-hot categorical
# df = pd.get_dummies(df, columns=cat_cols)
encoder = MultiColumnLabelEncoder(cat_cols)
encoder.fit(df)
# print(encoder.transform(df[cat_cols]))
df[cat_cols] = encoder.transform(df[cat_cols])

df
```

```python
# YOUR CODE
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


X = df.loc[:, df.columns!='Churn']
y = df["Churn"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9)
X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, train_size=0.5)

parameters = {
    'C': [100, 10, 1, 0.1, 0.01, 0.001],
    'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'max_iter': [500, 1000],
}

lr = LogisticRegression(random_state=42)

grid = GridSearchCV(lr, parameters, scoring='roc_auc', cv=5, n_jobs=-1)
grid.fit(X_train, y_train)

model = grid.best_estimator_

print(f"best score (train): {grid.best_score_}")
print(f"best model: {model}")
print(f"auc score (valid): {roc_auc_score(y_valid, model.predict_proba(X_valid)[:, 1])}")
print(f"auc score (test): {roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])}")
print()

# conf matrix
cnf_matrix = confusion_matrix(y_valid, model.predict(X_valid))
print("Confusion matrix:")
print(cnf_matrix)
print()

# classification report
clf_report = classification_report(y_valid, model.predict(X_valid))
print("Classification report:")
print(clf_report)
print()

y_pred_proba = model.predict_proba(X_valid)[:, 1]
fpr, tpr, _ = roc_curve(y_valid,  y_pred_proba)
auc = roc_auc_score(y_valid, y_pred_proba)
plt.plot(fpr, tpr, label=f"auc={str(auc)}")
plt.legend(loc=4)
plt.show()
```
